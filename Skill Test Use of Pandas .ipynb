```python
#Importing libraries
import pandas as pd
import numpy as np
```

```python
#Data Loading
df = pd.read_csv("Examen_mantenimiento.csv")
```

```python
#Variable types
print('Variable Types')
df.dtypes
```

```python
#Null Values
print('Null Values')
df.isnull().sum()
```

```python
#Dimensions
print('Shape:', df.shape)
df.shape
#Data Frame
print('Data first 5 rows')
df.head()
```

```python
df['Meses en mto']=df.iloc[:,4:].sum(axis=1)
df['Costo Total']=df['Meses en mto']*df['Costo']
df.head()
```

```python
#1.	¿Cuáles son los 5 proveedores que más cobraron en el año?
df.groupby('Proveedor').agg({'Costo Total' : 'sum'}).sort_values(by='Costo Total', ascending=False).head(5)
```

```python
#2.	¿Cuáles son los 5 proveedores que le dieron servicio a más tiendas en el año?
df.groupby('Proveedor').agg({'Tienda' : 'count'}).sort_values(by='Tienda', ascending=False).head(5)
```

```python
#3.	¿Cuál fue la tienda con mayor gasto de mantenimiento en marzo?
df_marzo = df[df['Marzo']==1]
df_marzo.groupby('Tienda').agg({'Costo' : 'sum'}).sort_values(by='Costo', ascending=False).head(1)
```

```python
#4.	¿Cuál fue la tienda con mayor gasto de mantenimiento en julio?
df_julio = df[df['Julio']==1]
df_julio.groupby('Tienda').agg({'Costo' : 'sum'}).sort_values(by='Costo', ascending=False).head(1)
```

```python
#5.	¿Cuál fue la tienda con mayor gasto de mantenimiento en todo el año?
df.groupby('Tienda').agg({'Costo Total' : 'sum'}).sort_values(by='Costo Total', ascending=False).head(1)
```

```python
#6.	¿Cuál es el mes en que se gastó más en mantenimiento a balanceadoras?
df.Equipo.unique()
```

```python
df_bal = df[df.Equipo == 'BALANCEADORA']
df_bal = df_bal.iloc[:,3:]
df_bal['Enero'] = df_bal['Costo']*df_bal['Enero']
df_bal['Febrero'] = df_bal['Costo']*df_bal['Febrero']
df_bal['Marzo'] = df_bal['Costo']*df_bal['Marzo']
df_bal['Abril'] = df_bal['Costo']*df_bal['Abril']
df_bal['Mayo'] = df_bal['Costo']*df_bal['Mayo']
df_bal['Junio'] = df_bal['Costo']*df_bal['Junio']
df_bal['Julio'] = df_bal['Costo']*df_bal['Julio']
df_bal['Agosto'] = df_bal['Costo']*df_bal['Agosto']
df_bal['Septiembre'] = df_bal['Costo']*df_bal['Septiembre']
df_bal['Octubre'] = df_bal['Costo']*df_bal['Octubre']
df_bal['Noviembre'] = df_bal['Costo']*df_bal['Noviembre']
df_bal['Diciembre'] = df_bal['Costo']*df_bal['Diciembre']
```

```python
bal_mes=df_bal.agg({'Enero': sum,'Febrero': sum,'Marzo': sum,'Abril': sum,'Mayo': sum,'Junio': sum,'Julio': sum,'Agosto': sum,'Septiembre': sum,'Octubre': sum,'Noviembre': sum,'Diciembre': sum})
```

```python
bal_mes.sort_values(ascending=False, inplace=True)
bal_mes.head(1)
```

```python
#7.	¿Cuánto se gastó por cada tipo de equipo en septiembre?
df_sept = df[df['Septiembre']==1]
df_sept.groupby('Equipo').agg({'Costo' : 'sum'}).sort_values(by='Costo', ascending=False)
```

```python
#8.	¿Cuánto se gastó por cada proveedor en mayo?
df_mayo = df[df['Mayo']==1]
df_mayo.groupby('Proveedor').agg({'Costo' : 'sum'}).sort_values(by='Costo', ascending=False)
```

```python
#9.	¿Cuánto se gastó por cada tipo de equipo en el año?
df.groupby('Equipo').agg({'Costo Total' : 'sum'}).sort_values(by='Costo Total', ascending=False)
```

```python
#10.	Genera una función para obtener el número total de servicios a los equipos por mes (la función tomará tres argumentos: el conjunto de datos, el equipo y el nombre o número del mes).
# Funcion num_servicios
# Calcula el número total de servicios por equipo en un mes específico
#
# Inputs:
# A. df = Dataframe con los datos a utilizar
# B. equipo = Entre comillas ('') y Mayúsculas, corresponde al tipo de equipo
#          i.e. 'BASCULA','BALANCEADORA','ANTENAS'
# C. mes = Entre comillas ('') y capitalizando la primer letra, corresponde al mes a analizar
#          i.e. 'Mayo', 'Enero', 'Diciembre'
#
# Output:
# Número de servicios por mes del equipo

def num_servicios(df, equipo, mes):
    df_equipo= df[df.Equipo== equipo]
    resp = df_equipo[mes].sum()
    return resp
    
```

```python
# Prueba
num_servicios(df,'BASCULA','Julio')
```
